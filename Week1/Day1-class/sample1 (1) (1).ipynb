```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import upper
spark = SparkSession.builder \
    .appName("Read File Example") \
    .getOrCreate()
```

```python
df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/workspace/default/practice-volume/C1.csv")
```

```python
df = spark.read.option("header", "true").csv("/Volumes/workspace/default/practice-volume/C1.csv")
```

```python
df.show()
```

```python
df = df.replace(["blank","Blank"],None)
```

```python
df = df.dropna()
```

```python
df = df.dropna(subset=["Sales","JoinDate","Category"])
```

```python
df = df.dropDuplicates()
```

```python
df.show()
```

```python
from pyspark.sql.functions import upper

df = df.withColumn("Country", upper(("Country")))
```

```python
df.show()
```
